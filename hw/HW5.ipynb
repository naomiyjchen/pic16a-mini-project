## Problem 1: Faceted Histogram

Run the following code block to define a function which generates two 1-dimensional `numpy` arrays. The first array, called `groups`, consists of random integers between `0` and `n_groups-1`, inclusive. The second array, called `data`, consists of real numbers. Each of these arrays has length `n`.   

````python
import numpy as np
from matplotlib import pyplot as plt

def create_data(n, n_groups):
    """
    generate a set of fake data with group labels. 
    n data points and group labels are generated. 
    n_groups controls the number of distinct groups. 
    Returns an np.array() of integer group labels and an 
    np.array() of float data. 
    """
    
    # random group assignments as integers between 0 and n_groups-1, inclusive
    groups = np.random.randint(0, n_groups, n)
    
    # function of the groups plus gaussian noise (bell curve)
    data  = np.sin(groups) + np.random.randn(n)
    
    return(groups, data)
````

### Requirements: 

1. Your function should work **whenever `m_rows*m_cols` is equal to the total number of groups.** Your function should first check that this is the case, and raise an informative `ValueError` if not. You may assume that there is at least one data point for each group label in the data supplied. 
2. You should include a horizontal axis label (of your choice) along **only the bottom row** of axes.

### Hints

- If your plots look "squished," then `plt.tight_layout()` is sometimes helpful. Just call it after constructing your figure, with no arguments. 
- Integer division `i // j` and remainders `i % j` are helpful here.  


````python
# your function facet_hist() here
def facet_hist(groups, data, m_rows, m_cols, figsize, **kwargs):
    '''
    This function facets the given data into groups and make separated histograms 
    accordingly.
    
    
    parameters
    __________
    groups: np.array of data
    data: the np.array of data as output by create_data().
    m_rows: the number of desired rows in your faceted histogram (explanation coming).
    m_cols: the number of desired columns in your faceted histogram (explanation coming).
    figsize: the size of the figure.
    **kwargs: parameters available in plt.hist()
    
    return
    ______
    a separate axis and histogram for each group
    '''
    n_groups = len(set(groups)) #total number of groups
    #if the number of histograms does not match the total number of groups
    if m_rows*m_cols != n_groups:
        #raise a ValueError
        raise ValueError('The number of histograms does not match the total number of groups.')
        
    fig, ax = plt.subplots(m_rows,m_cols) #create axes and figures
    plt.tight_layout() #adjust the amount of height and width reserved for space between subplots
        
    for r in range(m_rows):
        ax[r][0].set(ylabel = "Density") #set vertical axis labels along the leftmost column of axes
        
        for c in range(m_cols):
            ax[m_rows-1][c].set(xlabel = "x") # set horizontal axis labels along the bottom row of axes
            group_number = m_rows*c+r # group number 
            ax[r][c].set(title = "Group " + str(group_number)) #set titles for each histogram
            
            x_group = data[groups == group_number] # extract data of one group
            ax[r][c].hist(x_group,**kwargs) #plot the histogram
        
    
    
    
````

There is a separate scatterplot for each possible pair of variables. In fact, there are two: one where the first variable is on the horizontal axis, and one where it's on the vertical axis. Some analysts prefer to remove half the plots to avoid redundancy, but you don't have to bother with that. The diagonal is empty, since there's no point in investigating the relationship between a variable and itself. 

Don't forget comments and docstrings! 

````python
# your function scatterplot_matrix here
def scatterplot_matrix(cols,figsize):
    '''
    This functions generates scatter plots of all possible pairs of variables in the cols array
    
    '''
    n_cols = len(cols) #number of variables
    fig, ax = plt.subplots(n_cols, n_cols, figsize = figsize) # create empty figures
    plt.subplots_adjust(wspace =0.5,hspace=0.5)  #adjust the amount of height and width reserved for space between subplots
    data = happiness[cols] #extract data needed
    
    for r in range(n_cols):
        for c in range(n_cols):
            ax[r][c].set(title = cols[r], 
                         ylabel = cols[c])
            
            if r != c: #avoid plotting in the diagonal of the matrix
                ax[r][c].scatter(data[cols[r]],data[cols[c]])
                
                corr = np.corrcoef(data[cols[r]],data[cols[c]]) #compute correlation of two variables
                ax[r][c].set(xlabel = r"$\rho$ = " + str(np.round(corr[0][1], 2))) #set the correlation as xlabel
        
````
